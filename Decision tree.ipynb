```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz,plot_tree
from IPython.display import Image
```

```python
df = pd.read_csv("/content/demoDT.csv",sep=",")
print(df)
```

```python
feature_cols = ['Literacy',' Cleanliness',' Crime_Rate']
feature = df[feature_cols]
Target = df.good
```

```python
DT = DecisionTreeClassifier()
DT = DT.fit(feature,Target)
```

```python
l = int(input("Enter Literacy:"))
cl = int(input("Enter Cleanliness:"))
cr = int(input("Enter CrimeRate:"))
pred = DT.predict([[l,cl,cr]])
if pred ==1:
  print("good state")
else:
    print("bad state")
```

```python
plot_tree(DT)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier

```

```python
df = pd.read_csv('/content/demoDT.csv',sep=",")
print(df)
```

```python
feature_cols = ['Literacy',' Cleanliness',' Crime_Rate']
feature = df[feature_cols]
Target = df.good
```

```python
RF = RandomForestClassifier(n_estimators=100)
RF.fit(feature,Target)
```

```python
l = int(input("Enter Literacy:"))
cl = int(input("Enter Cleanliness:"))
cr = int(input("Enter CrimeRate:"))
pred = RF.predict([[l,cl,cr]])
if pred ==1:
  print("good state")
else:
    print("bad state")
```
